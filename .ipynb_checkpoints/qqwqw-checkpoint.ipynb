```python
def get_spectrum_ABS(sig, fd, N):
    spectrum = np.abs(np.fft.rfft(sig,2**13,norm='forward'))*2
    frequency_array = np.fft.rfftfreq(len(sig), 1./fd)
    spectrum[0] = spectrum[0]/2
    return spectrum, frequency_array
```

```python
def get_amplitude_correction(freq,fd,N):
    pure_sig = np.array([1.*math.sin(2.*np.pi*freq*t/fd) for t in range(N)])
    spectrum, frequency_array = get_spectrum_ABS(pure_sig, fd, N)
    peaks_index, _ = find_peaks(spectrum, height=0.1)
    q = 1/spectrum[peaks_index]
    if freq == 0: q = 1
    return q
```

```python
def get_list_of_freq(max_freq,n):
    return np.linspace(0,max_freq,n*max_freq+1)
```

```python
data = np.loadtxt("sig.txt", dtype=int, skiprows=0, usecols=1)
print('len:',len(data))



amplitude_convert = 330 # конвертация амплитуды в мВ
fd = 409  # частота дискретизации
N = len(data) # количество отсчетов в записи
T = N / fd # период дескретизации
max_freq = round(fd/2) # максимальтная частота в спектре
n = 2 # делитель для привязки частот (на сколько частей делится промежуток в 1 Гц)
detect_height = 0.1 # мВ уровень определения пиков

# Конвертация амплитуды в мВ
data = data/amplitude_convert

plt.plot(data)
plt.grid(True)
```

```python
#Выделение участка записи
#45471 53010
#53738 67137
begin_index = 53738
end_index = 67137
print('begin/end index:',begin_index,'-',end_index)

sig_window = data[begin_index: end_index]
N = len(sig_window)
print('len:',N)

plt.plot(sig_window[0:100])
plt.grid(True)
```

```python
# Получение модуля амплитудного спектра сигнала 
# перевод амплитуды в размах
spectrum, frequency_array = get_spectrum_ABS(sig_window, fd, N)
spectrum = spectrum*2

# Нахождение пиков спектра
detect_height = 1
peaks_index, _ = find_peaks(spectrum, height=detect_height)
print('Найденные амплитуда и частота пика:      ',np.around(spectrum[peaks_index],3), np.around(frequency_array[peaks_index],2))

# Вывод графика спектра 
plt.plot(frequency_array,spectrum)
plt.grid(True)
plt.plot(frequency_array[peaks_index],spectrum[peaks_index],'x')
for index in peaks_index:
    plt.text(frequency_array[index], spectrum[index], str(round(spectrum[index],2))+'mV, ' + str(round(frequency_array[index],3))+ ' Hz', size=14)
plt.show()
```

```python
# # Нормализация амплитуд 
list_of_freq = get_list_of_freq(max_freq, n)

for index in peaks_index:
    frec_normalize = list_of_freq[np.abs(list_of_freq - frequency_array[index]).argmin()]
    amplitude_correction = get_amplitude_correction(frec_normalize,fd,N)
    print('Коррекция амплитуды для частоты',str(round(frequency_array[index],2))+':',amplitude_correction)
    spectrum[index] = spectrum[index] * amplitude_correction
```

```python
# Вывод графика спектра с нормализацией
plt.plot(frequency_array,spectrum)
plt.grid(True)
plt.plot(frequency_array[peaks_index],spectrum[peaks_index],'x')
for index in peaks_index:
    plt.text(frequency_array[index], spectrum[index], str(round(spectrum[index],2))+'mV, ' + str(round(frequency_array[index],3))+ ' Hz', size=14)
plt.show()
```
